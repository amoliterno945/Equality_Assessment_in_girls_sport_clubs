```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
council=pd.read_csv("council_district_info.csv")
GPLA=pd.read_csv("GPLA_Report.csv")

#Clean
GPLA.drop('Unnamed: 14',axis=1,inplace=True)
GPLA.drop(258,axis=0,inplace=True)

#Fix typo
for i in range(GPLA.shape[0]):
    if GPLA.loc[i,'Type'] == "Basketbll":
        GPLA.loc[i,'Type'] = 'Basketball'

#Fill Season's NAs with the season mentioned in Description
for i in GPLA.loc[GPLA.Season.isnull(),['Short Description']].index:
    GPLA.loc[i,['Season']] = GPLA.loc[i,['Short Description']].values[0].split(" ",1)[0]

#Join tables by District
dat=GPLA.merge(council, how='left',left_on='Council District', right_on='District')
dat=dat.sort_values(by='Enrolled')
dat.reset_index(inplace=True,drop=True)

#Add a single column for hours
dat.loc[:,"Begin Time"]=[i.strip().lower() for i in dat.loc[:,"Begin Time"]]
dat.loc[:,"End Time"]=[i.strip().lower() for i in dat.loc[:,"End Time"]]
dat.loc[:,"Timespan"] = dat.loc[:,"Begin Time"]+"-"+dat.loc[:,"End Time"]
dat.loc[:,"hours"] = (pd.to_datetime(dat.loc[:,"End Time"])-\
                      pd.to_datetime(dat.loc[:,"Begin Time"])).astype('timedelta64[h]')
```

```python
from matplotlib import cm
from matplotlib.cm import ScalarMappable

#Normalize wages t0 0-1 for colormapping
grouped_mean=dat.groupby('Council District').Avg_wage.mean()
group_min=np.min(dat.groupby('Council District').Avg_wage.mean())
norm=grouped_mean-group_min
norm=norm/np.max(norm)

#Plot bar chart with colors as 3rd dimension
plt.figure(figsize=(10,7))
plt.bar(x=dat.groupby('Council District').Enrolled.sum().index,
        height=dat.groupby('Council District').Enrolled.sum(),
        color=cm.viridis(norm))
plt.xticks(dat.groupby('Council District').Enrolled.sum().index)
plt.title("District Enrollment Count")

#Colorbar to the side
sm = ScalarMappable(cmap=cm.viridis)
sm.set_array([])
plt.colorbar(sm)

corr=np.corrcoef(dat.groupby('Council District').Enrolled.sum(),norm)[0][1]
print("Correlation of Income vs. Enrollment: ",np.round(corr,2))
```

```python
#Age inclusion

d = dict.fromkeys(np.arange(4,18), 0)

for i in d.keys():
    for j in range(dat.shape[0]):
        if i >= dat.loc[j,'Begin Age'] and i <= dat.loc[j,'End Age']:
            d[i] = d[i] + dat.loc[j,'Enrolled']

plt.figure(figsize=(10,7))
plt.title("Age Inclusion")
plt.bar(d.keys(),d.values())
```

```python
timespans={}
for i in dat.Timespan.value_counts().index:
    timespans[i]={}

for i in list(timespans.keys()):
    timespans[i]['Clubs'] = dat.Timespan.value_counts()[i]
    timespans[i]['Enrolled'] = dat.groupby('Timespan').Enrolled.agg(sum)[i]
    timespans[i]['Hours'] = dat.loc[dat.Timespan==i,'hours'].values[0]
    
timespans=pd.DataFrame(timespans).T
```

```python
timespans.sort_values('Enrolled',ascending=False)
```

```python
fig, ax  = plt.subplots(nrows=1, ncols= 2,figsize=(10,4)) 

ax[0].scatter(timespans.Clubs,timespans.Enrolled)
ax[0].set_title('Number of Clubs vs. Enrollment')
ax[1].scatter(timespans.Hours,timespans.Enrolled)
ax[1].set_title('Hours vs. Enrollment')


ax[0].set(xlabel="Clubs", ylabel="Enrollment")
ax[1].set(xlabel="Hours", ylabel="Enrollment")
```
